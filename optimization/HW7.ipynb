## Question 2

Let $f(x_{1},x_{2})=x_{1}^{2}+x_{2}^{2}-x_{1}x_{2}-2x_{1}$.  Perform the backtracking procedure at $x=(1,1)$ in a direction $d=(2,-1)$ with $s=1, \alpha=0.7,\beta=0.5$. 

$d$ is a descent direction for the function, proved in (q1)

$\alpha \nabla f(x)^{T}d$= $-0.7 \left[ \begin{array}{cc} -1 & 1 \end{array} \right] \left[ \begin{array}{c} 2 \\ -1 \end{array} \right] = -.7*3 = -2.1 $, so while $\frac{f(x+td)-f(x)}{t}<-2.1$ we continue dividing $t$ by $\beta$ until we find a value of t that satisfies the stopping condition

$$\begin{array}{c|c|c|c|c|c} t & x+td & f(x+td) & f(x) & \frac{f(x+td)-f(x)}{t}  \\ \hline \\ 1 & (3,0) & 3 & -1 & 4 \\ \frac{1}{2} & (2,\frac{1}{2}) & -\frac{3}{4} & -1 & \frac{1}{2} \\ \frac{1}{4} & (\frac{3}{2},\frac{3}{4}) & -\frac{21}{16}  & -1 & -\frac{5}{4} \\ \frac{1}{8} & (\frac{5}{4},\frac{7}{8}) & -\frac{81}{64} & -1 & -\frac{17}{8}\end{array}$$

As $-\frac{17}{8}=-2.125<-2.1$ we can stop the backtracking procedure and our step size is .125

## Question 3

$f(x)=\sum_{i=1}^{m}(||x-a_{i}||-d_{i})^{2}$

## Part A

$\nabla f(x)= \sum_{i=1}^{m}2(||x-a_{i}||-d_{i})\frac{x-a_{i}}{||x-a_{i}||}$

## Part B

$\nabla f(x)=0$, when $0=\sum_{i=1}^{m}2(||x-a_{i}||-d_{i})\frac{x-a_{i}}{||x-a_{i}||}$

$\sum_{i=1}^{m}2(||x-a_{i}||-d_{i})\frac{x-a_{i}}{||x-a_{i}||}=0 \Leftrightarrow \sum_{i=1}^{m}(x-a_{i}-d_{i}\frac{x-a_{i}}{||x-a_{i}||})=0 $

$\Leftrightarrow 0 = \sum_{i=1}^{m}x - \sum_{i=1}^{m}a_{i}-\sum_{i=1}^{m}(d_{i}\frac{x-a_{i}}{||x-a_{i}||})$

$\Leftrightarrow mx = \sum_{i=1}^{m}a_{i} + \sum_{i=1}^{m}(d_{i}\frac{x-a_{i}}{||x-a_{i}||}) \Leftrightarrow x = \frac{1}{m}(\sum_{i=1}^{m}a_{i} + \sum_{i=1}^{m}(d_{i}\frac{x-a_{i}}{||x-a_{i}||})$

## Part C

$0=2x\sum_{i=1}^{m}\frac{||x-a_{i}||-d_{i}}{||x-a_{i}||} - 2\sum_{i=1}^{m}\frac{a_{i}(||x-a_{i}||-d_{i})}{||x-a_{i}||}$

$2\sum_{i=1}^{m}\frac{a_{i}(||x-a_{i}||-d_{i})}{||x-a_{i}||}=2x\sum_{i=1}^{m}\frac{||x-a_{i}||-d_{i}}{||x-a_{i}||}$ we can divide as we assume that all $x_{k}$s are different from $a_{i}$

$x=\frac{1}{\sum_{i=1}^{m}\frac{||x-a_{i}||-d_{i}}{||x-a_{i}||}}\sum_{i=1}^{m}\frac{a_{i}(||x-a_{i}||-d_{i})}{||x-a_{i}||}$

$T(x)=\frac{1}{\sum_{i=1}^{m}\frac{||x-a_{i}||-d_{i}}{||x-a_{i}||}}\sum_{i=1}^{m}\frac{a_{i}(||x-a_{i}||-d_{i})}{||x-a_{i}||}$

$x_{k+1} = T(x_{k})$

$x_{k+1} = \frac{1}{\sum_{i=1}^{m}\frac{||x_{k}-a_{i}||-d_{i}}{||x_{k}-a_{i}||}}\sum_{i=1}^{m}\frac{a_{i}(||x_{k}-a_{i}||-d_{i})}{||x_{k}-a_{i}||}$

$x_{k+1}=x_{k}-x_{k}+\frac{1}{\sum_{i=1}^{m}\frac{||x_{k}-a_{i}||-d_{i}}{||x_{k}-a_{i}||}}\sum_{i=1}^{m}\frac{a_{i}(||x_{k}-a_{i}||-d_{i})}{||x_{k}-a_{i}||}$

$x_{k+1}=x_{k}-\frac{1}{\sum_{i=1}^{m}\frac{||x_{k}-a_{i}||-d_{i}}{||x_{k}-a_{i}||}}\sum_{i=1}^{m}\frac{(x_{k}-a_{i})(||x_{k}-a_{i}||-d_{i})}{||x_{k}-a_{i}||}$

$=x_{k}-\frac{1}{\sum_{i=1}^{m}\frac{||x_{k}-a_{i}||-d_{i}}{||x_{k}-a_{i}||}}\nabla f(x_{k})$

Therefore we have a gradient method with a step size of $\frac{1}{\sum_{i=1}^{m}\frac{||x_{k}-a_{i}||-d_{i}}{||x_{k}-a_{i}||}}$ as it is set up in the form of a gradient method

## Question 4

$min_{x\in \mathbb{R}^{3}} f(x)=\sum_{i=1}^{9}\omega_{i}||x-a_{i}||$

$\nabla f(x)=\sum_{i=1}^{9}\omega_{i}\frac{x-a_{i}}{||x-a_{i}||}$

$x_{k+1}=\frac{1}{\sum_{i=1}^{9}\frac{\omega_{i}}{||x-a_{i}||}}\sum_{i=1}^{9}\frac{\omega_{i}a_{i}}{||x-a_{i}||}$

```julia
using LinearAlgebra
using PyPlot
```

```julia
a = [-10 10 0; 0 30 -10; 20 20 10; 30 0 5; 25 -5 20; -20 -25 5; 30 25 -10; -20 20 10; 0 -15 0]
```

```julia
scatter3D(a[:,1],a[:,2],a[:,3]);
```

```julia
omega = [1;2;.5;.5;1/3;1;1.5;1;1]
```

```julia
x=[0 0 0]
```

```julia
function nextX(x,omega,alpha)
    s1 = [0 0 0]
    s2 = 0
    for i=1;9
        s1 = s1 .+ omega[i]/norm(x.-alpha[i,:])*alpha[i]/norm(x.-alpha[i,:]);
        s2 = s2 + omega[i]/norm(x.-alpha[i,:]);
    end
    return 1/s2 * s1;
end
```

```julia
b = nextX(x,omega,a)
```

This proccess can be continued until $\nabla f(x)$ is close to zero, here it is (-.8,-.8,-.8), so we continue

```julia
c = nextX(b,omega,a)
```

We continue as $\nabla f(x)$ is (-0.16,-0.16,-0.16)

```julia
d = nextX(c,omega,a)
```

We continue as $\nabla f(x)$ is -.003 (1,1,1), however that is rather small so continuing more will only generate smaller and smaller returns

```julia
e = nextX(d,omega,a)
```

```julia
f = nextX(e,omega,a)
```

```julia
g = (0,0,0)
for i=1:1000
    g = nextX(f,omega,a);
end
```

```julia
g
```

During the next 1000 increments, our solution only improves by roughly -0.00051.  As such we could've been fine with f, or even e, or d depending on accuracy wanted
