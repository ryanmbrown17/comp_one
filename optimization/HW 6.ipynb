## Question 1

```julia
x_k = [-4;-3.5;-3;-2.5;1;1.2;1.5;1.7;2]
```

```julia
y_k = [-3;-2;-3;-4;-3;1;0;-3;-5];
```

```julia
figure(figsize=(10,4));
scatter(x_k,y_k);
grid();
```

Quadratic Form 
$$f(x)=ax^{2}+bx+c$$
Error
$$\varphi(\alpha)=||A\alpha-B||^{2}$$

```julia
A = [x_k.^2 x_k ones(9)]
```

```julia
alpha = A\y_k
```

```julia
t = range(-4, length=100, stop=2)
```

```julia
ft = alpha[1]*t.^2 .+ alpha[2]*t .+ alpha[3];
```

```julia
figure(figsize=(10,4));
scatter(x_k,y_k);
plot(t,ft);
grid();
```

Polynomial of Degree 8
$$f(x)=ax^{8}+bx^{7}+cx^{6}+dx^{5}+ex^{4}+fx^{3}+gx^{2}+hx+i $$
Error
$$\varphi(\alpha)=||A\alpha-B||^{2}$$

```julia
A2 = [x_k.^8 x_k.^7 x_k.^6 x_k.^5 x_k.^4 x_k.^3 x_k.^2 x_k ones(9)]
```

```julia
alpha2 = A2\y_k
```

```julia
ft2 = [t.^8 t.^7 t.^6 t.^5 t.^4 t.^3 t.^2 t ones(100)]*alpha2;
```

```julia
figure(figsize=(10,4));
scatter(x_k,y_k);
plot(t,ft2);
grid();
```

While the polynomial of degree 8 does go through every point, it would be unlikely that a set of data would include the large dip between x=-2.5 and x=1 that occurs in the polynomial.  For that reason the quadratic would be a batter choice for this particular data set

## Question 3

```julia
n=100;
t=sort(rand(n));
b=(t.-0.3).*sin.(pi*t.^2)-0.2*rand(n);
```

```julia
figure(figsize=(10,4));
scatter(t,b);
axis("equal");
grid();
```

$$min_{x\in R^{n}}||x-b||^{2}+\lambda R(x) $$
$$R(x)=\Sigma_{i=1}^{n-1}(x_{i}-x_{i=1})^{2}$$

```julia
D = diagm(0 => ones(n), 1 =>-ones(n));
```

```julia
D = D[1:end-1, 1:end-1];
```

```julia
lambda = 5;
xLS = (I+lambda*D'*D)\b;
lambda = 100;
xLS2 = (I+lambda*D'*D)\b;
lambda = 500;
xLS3 = (I+lambda*D'*D)\b;
t_norm = range(0,stop=1,length=100);
```

```julia
figure(figsize=(10,4));
grid();
#plot(t,x);
scatter(t,b, s=2, c="red");
scatter(t_norm,xLS, s=2, c="green");
scatter(t_norm,xLS2, s=2, c="#0000FF");
scatter(t_norm,xLS3, s=2, c="Orange");
```

##  Question 4

```julia
m=50;
alpha = -rand(m);
beta = rand(m) + ones(m);
eta = 2*rand(m) .+ 3*ones(m);
theta = 2pi*rand(m);
```

$$(u_{i},v_{i})=(\alpha_{i} + \eta_{i}cos\theta_{i}+\eta_nsin\theta_{i})$$

```julia
u = alpha .+ eta.*cos.(theta);
v = beta  .+ eta.*sin.(theta);
```

```julia
figure(figsize=(10,4));
scatter(u,v);
axis("equal");
grid();
```

```julia
uv = [u v];
```

```julia
A = [-2uv ones(m)];
```

```julia
y = -u.^2 -v.^2;
```

```julia
xhat = A\y
```

# Center

```julia
x = xhat[1:end-1]
```

## Radius

```julia
R = xhat[end]
```

```julia
r = sqrt(norm(x)^2-R)
```

## Graph

```julia
theta_graph =  range(0, stop=2pi, length=100);
```

```julia
cx = x[1] .+ r*cos.(theta_graph);
cy = x[2] .+ r*sin.(theta_graph);
```

```julia
figure(figsize=(4,4));
grid();
axis("equal");
scatter(u, v);
plot(cx,cy, "g");
```
